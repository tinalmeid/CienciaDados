Resposta da Atividade da Aula Prática da Unidade 2

```python
#classe para inserção do livro
class Livro:
  # _init_ construtor da classe (padrão)
  #self inicializa os atributos da classe (padrão)
  # titulo, autor, genero e qtde são meus atrubutos para a classe livro
  def __init__(self, titulo, autor, genero, qtde):
    self.titulo = titulo
    self.autor = autor
    self.genero =  genero
    self.qtde = qtde

  # A saída do meu método é do tipo String, com todos os parametros da classe
  def __str__(self):
    return f'Título: {self.titulo}, Autor: {self.autor}, Gênero: {self.genero}, Quantidade: {self.qtde}'
```

```python
#criando uma lista de livros
minha_biblioteca = []
#criando lista para armazenar genero
genero_livro = []
#criando lista para armazenar a quantidade disponível
quantidade = []
```

Gerenciando meus livros

```python
#função para cadastrar livros
#usando os parâmetros da minha classe Livro
def adicionar_livro(titulo, autor, genero, qtde):
  #adicionando um novo livro
  novo_livro = Livro(titulo, autor, genero, qtde)
  #inclui o livro na minha biblioteca
  minha_biblioteca.append(novo_livro)
  #adicionando o genero na minha lista genero
  genero_livro.append(genero)
  #adicionando a quantidade na minha lista quantidade
  quantidade.append(qtde)
  print('O Livro', titulo, ',de gênero: ',genero, 'foi adicionado com sucesso na sua biblioteca!')

#adicionando livros
adicionar_livro(titulo="Mindset: A nova psicologia do sucesso" ,autor="Carol S. Dweck", genero="Psicologia", qtde="15")
adicionar_livro(titulo="Hábitos Atômicos: um Método Fácil de Criar Bons Hábitos",autor="James Clear", genero="Desenvolvimento Humano", qtde="13")
adicionar_livro(titulo="O Projeto Fênix: um Romance Sobre TI",autor="Gene Kim, Kevin Behr, George Spafford", genero="Romance", qtde="21")
adicionar_livro(titulo="Código Limpo: Habilidades Práticas do Agile Software",autor="Robert C. Martin", genero="Tecnologia", qtde="6")
adicionar_livro(titulo="Inspirado: Como Criar Produtos de Tecnologia que os Clientes Amam",autor="Marty Cagan", genero="Gerência de Projetos", qtde="16")
```

```python
#função para listar os livros cadastrados
def listar_livros():
  print('A Minha biblioteca possui os seguintes livros:')
  for livro in minha_biblioteca:
    print(livro)


#chama a função de listar os livros
listar_livros()
```

```python
# Função para buscar livros pelo título
def buscar_livros():
    #input para o usuário digitar o título desejado
    #usando o para busca de maneira case-insensitive.
    busca = input("Digite o título (ou parte) do livro que deseja procurar: ").lower()
    #vai percorrer todos os livros da biblioteca de maneira case-insensitive
    #se encontar a palvra que está buscando, armazena o livro encontrado
    #na lista livros_encotrado
    livros_encontrados = [livro for livro in minha_biblioteca if busca in livro.titulo.lower()]

    if livros_encontrados:
        print(f"Livros encontrados com a palavra '{busca}':")
        for livro in livros_encontrados:
            #Exibe o livro encontrdo
            print(livro)
    #se o livro não existir
    else:
        print(f"Nenhum livro encontrado com a palavra '{busca}'.")

#chamando a função de buscar livro
buscar_livros()
```

```python
#importando a biblioteca para a geração dos gráficos
import matplotlib.pyplot as plt
#Gerando o gráfico em linhas
plt.plot(genero_livro, quantidade, marker='o', linestyle='-')

# Colocando os legendas dos rótulos e o título no gráfico e Colocando os legendas na diagonal
plt.xlabel('Gênero')
plt.ylabel('Quantidade')
plt.title('Indicadores da Quantidade de Livros por Gênero')
plt.xticks(rotation=45)

#adicionando os rótulos nos pontos dos dados
for i, valor in enumerate(quantidade):
    plt.text(i, valor, str(valor), ha='center', va='bottom')

#mostrando as linhas de grade
plt.grid(True)

#o gráfico
plt.show()
```
